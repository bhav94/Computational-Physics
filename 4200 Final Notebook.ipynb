# The Simple Harmonic Oscillator (Ordinary Differential Equations)


Following along with the math on this website http://farside.ph.utexas.edu/teaching/315/Waves/node10.html 

We want to try and solve the second order differential equation for a damped harmonic oscillator. Here Newton's second law takes the form:  

$$ ma + cv + kx = 0 $$ 

where *a* is acceleration, *c* is a viscous damping proportionality constant, and *k* is the spring constant. So Newton's second law can be rewritten as:

$$ m\frac{d^{2}x}{dt^{2}} + c\frac{dx}{dt} + kt = 0 $$

Dividing out the *m* we get,

$$ \ddot{x} + \frac{c}{m}\dot{x} + \omega^{2}x = 0 $$

where $\omega = \sqrt{\frac{k}{m}}$ is the undamped oscillation frequency. 


```python
import numpy as np
from scipy.integrate import odeint
from matplotlib import pyplot as plt
%matplotlib inline
```

We define our constants present in the above second order differential (mass *m*, spring constant *k*, and the viscous constant *c*) with the appropriate units shown. 

```python
m = 0.5  # mass in kg
k = 4  # spring constant in N/m
visc = 0.5  # viscosity constant in N*s/m
```

If we guess a solution to the differential equation to be of the form 

$$ x(t) = ae^{-\gamma t} \cos{(\omega_{1}t - \phi)} $$

we can plug it back into our second order diff eq and eventually see that $ \gamma = \frac{c}{2m}$ and $ \omega_{1} = \sqrt{\omega^{2} - \frac{c^{2}}{4m^{2}}}$

```python
omega = np.sqrt(k / m) #undamped frequency 
gamma = visc / (2 * m) #gamma from above equation
```

```python
#defining how many steps calculation will take
#numerous steps are taken here to ensure a smooth curve, fewer time steps result in a triangular shape instead.
time = np.linspace(0, 10, 1000)

#initial condition, which is mass's starting position
y_initial = (3, 0)
```

Followning along with the math here: https://faculty1.coloradocollege.edu/~sburns/toolbox/ODE_II.html

Differential equations look like
$$ \frac{d\mathbf{X}}{dt} = \mathbf{f}(\mathbf{X}, t) $$

where $\mathbf{X}$ is a vector of parameters $[x(t), y(t), v_{x}(t), v_{y}(t),...]$ and $\mathbf{f}$ is a vector of corresponding parameters $[v_{x}(t), v_{y}(t), \frac{F_{x}}{m}, \frac{F_{y}}{m},...] $.

So for a damped harmonic oscillator with an equation like the one we described above:

$$\mathbf{X} = \big[ x , v \big] $$
and
$$\mathbf{f} = \big[ v , -\frac{k}{m}x - \frac{c}{m}v \big] $$

So *yvec* is our $\mathbf{X}$ and the function defined below,  *deriv*, is our $\mathbf{f}$. As we can see below, it is returning *yvec[1]*, which is the velocity ($\dot{x}$), and the expression afterwards is what $\ddot{x}$ equals. 

```python
def deriv(yvec, time, gamma, omega):
    return (yvec[1], -2 * gamma * yvec[1] - omega **2 * yvec[0])
```

```python
#actually integrating to find the solution to our differential equation!
#deriv function passed as first argument, intial conditions given, time given, and extra arguments given or else it raises an error message

sol = odeint(deriv, y_initial, time, args=(gamma, omega))
```

```python
#looking at output 
sol
```

We see that the output of the odeint function, *sol* is an array. There are two columns, the first is the position (y) for every time t. The second column is the velocity (y') for corresponding times. This checks out as the first row represents t=0 and we see (position, velocity) = (3, 0).

```python
plt.figure(figsize=(10, 10))

plt.plot(time, sol[:, 0], label='y')
plt.plot(time, sol[:, 1], label="y'")

plt.title('Damped Harmonic Oscillator', fontsize=20)
plt.ylabel('Amplitude (m)', fontsize=20)
plt.xlabel('Time (s)', fontsize=20)

plt.yticks(fontsize=15)
plt.xticks(fontsize=15)

plt.legend(loc='best', fontsize=20)
plt.show()
```

These results line up with what we would expect from a damped harmonic oscillator. We see that with the constants that we have now, the system is damped but it's not quite overdamped. If we were to drastically increase the frictional component (*visc*) we would see the exponential dropoff characteristic of overdampening. 

```python
m = 0.5  # mass in kg
k = 4  # spring constant in N/m
visc = 3  # viscosity constant in N*s/m

omega = np.sqrt(k / m) #undamped frequency 
gamma = visc / (2 * m) #gamma from above equation

#defining how many steps calculation will take
#numerous steps are taken here to ensure a smooth curve, fewer time steps result in a triangular shape instead.
time = np.linspace(0, 10, 1000)

#initial condition, which is mass's starting position
y_initial = (3, 0)

def deriv(yvec, time, gamma, omega):
    return (yvec[1], -2 * gamma * yvec[1] - omega **2 * yvec[0])

sol = odeint(deriv, y_initial, time, args=(gamma, omega))

plt.figure(figsize=(10, 10))

plt.plot(time, sol[:, 0], label='y')
plt.plot(time, sol[:, 1], label="y'")

plt.title('Overdamped Harmonic Oscillator', fontsize=20)
plt.ylabel('Amplitude (m)', fontsize=20)
plt.xlabel('Time (s)', fontsize=20)

plt.yticks(fontsize=15)
plt.xticks(fontsize=15)

plt.legend(loc='best', fontsize=20)
plt.show()
```

If we were to get rid of the frictional constant entirely (setting it equal to zero), we would expect perpetual harmonic motion. The graph would be of a cosine wave and it would never decrease in amplitude. 

We can test that by increasing time to 40 seconds (just to look further ahead) and setting visc = 0.

```python
m = 0.5  # mass in kg
k = 4  # spring constant in N/m
visc = 0  # viscosity constant in N*s/m

omega = np.sqrt(k / m) #undamped frequency 
gamma = visc / (2 * m) #gamma from above equation

#defining how many steps calculation will take
#numerous steps are taken here to ensure a smooth curve, fewer time steps result in a triangular shape instead.
time = np.linspace(0, 40, 1000)

#initial condition, which is mass's starting position
y_initial = (3, 0)

def deriv(yvec, time, gamma, omega):
    return (yvec[1], -2 * gamma * yvec[1] - omega **2 * yvec[0])

sol = odeint(deriv, y_initial, time, args=(gamma, omega))

plt.figure(figsize=(10, 10))

plt.plot(time, sol[:, 0], label='y')
plt.plot(time, sol[:, 1], label="y'")

plt.title('Simple Harmonic Oscillator', fontsize=20)
plt.ylabel('Amplitude (m)', fontsize=20)
plt.xlabel('Time (s)', fontsize=20)

plt.yticks(fontsize=15)
plt.xticks(fontsize=15)

plt.legend(loc='best', fontsize=20)
plt.show()
```
