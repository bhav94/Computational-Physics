# Fitting Data


#### Example: Temperature Calibration

This notebook will be an exercise in making best fit lines with different sets of data. 

We will use temperature and voltage as our parameters of data, but we will change the temperature array values to see the effects on the best fit line. 

### Importing necessary packages

```python
import matplotlib.pyplot as plt
from matplotlib import pylab

import numpy as np
from scipy import stats

%matplotlib inline
```

### Creating data for temperature (units of Kelvin) and Voltage (units of volts).  

```python
Voltage_V = np.array([1.1,1.2,1.30,1.40,1.50,1.60,1.70,1.80,1.90])
Temperature_K1 = np.array([190, 200, 205, 215, 220, 230, 235, 255, 240]) 
#called Temperature_K1 because we will be using different sets of temperatures
#to plot

print('Voltage [V]: ', Voltage_V)
print('Temperature [K]: ', Temperature_K1)
```

```python
# Generating a linear fit

slope1, intercept1, r_value1, p_value1, std_err1 = stats.linregress(Voltage_V,Temperature_K1)
line_fit1 = slope1 * Voltage_V + intercept1
```

```python
# Getting important information about data set, necessary to graph

print('Slope : ',slope1)
print('Intercept : ', intercept1)
print('R_value : ', r_value1)
print('p_value : ', p_value1)
print('std_error : ', std_err1)
```

### Creating array of values for Temperature_K2 (changing some temp values)

```python
Voltage_V = np.array([1.1,1.2,1.30,1.40,1.50,1.60,1.70,1.80,1.90])
Temperature_K2 = np.array([190, 215, 226, 224, 235, 250, 260, 272, 280])

print('Voltage [V]: ', Voltage_V)
print('Temperature [K]: ', Temperature_K2)
```

```python
# Generating a linear fit

slope2, intercept2, r_value2, p_value2, std_err2 = stats.linregress(Voltage_V,Temperature_K2)
line_fit2 = slope2 * Voltage_V + intercept2
```

```python
print('Slope : ',slope2)
print('Intercept : ', intercept2)
print('R_value : ', r_value2)
print('p_value : ', p_value2)
print('std_error : ', std_err2)
```

### Temperature_K3 (changing the array once more to get a third set of different data)

```python
Voltage_V = np.array([1.1,1.2,1.30,1.40,1.50,1.60,1.70,1.80,1.90])
Temperature_K3 = np.array([210, 190, 220, 235, 243, 307, 245, 261 , 278])

print('Voltage [V]: ', Voltage_V)
print('Temperature [K]: ', Temperature_K3)
```

```python
# Generated linear fit
slope3, intercept3, r_value3, p_value3, std_err3 = stats.linregress(Voltage_V,Temperature_K3)
line_fit3 = slope3 * Voltage_V + intercept3
```

```python
print('Slope : ',slope3)
print('Intercept : ', intercept3)
print('R_value : ', r_value3)
print('p_value : ', p_value3)
print('std_error : ', std_err3)
```

### Calibration

```python
#randomly measured a voltage of 2.0352 V and figuring out what temperature it corresponds to

New_Voltage_V = 2.0352
New_Temp_K1 = New_Voltage_V * slope1
New_Temp_K2 = New_Voltage_V * slope2
New_Temp_K3 = New_Voltage_V * slope3

print('New Temperature Measurement 1 [K] :',New_Temp_K1)
print('New Temperature Measurement 2 [K] :',New_Temp_K2)
print('New Temperature Measurement 3 [K] :',New_Temp_K3)
```

### Plotting Data vs. Fit

```python
plt.figure(figsize=(20,10))

plt.plot(Voltage_V,Temperature_K1,'b*', label="Temperature Sensor #1")
plt.plot(Voltage_V, line_fit1,'b-',label="")

plt.plot(Voltage_V,Temperature_K2,'rD', label="Temperature Sensor #2")
plt.plot(Voltage_V, line_fit2,'r-',label="")

plt.plot(Voltage_V,Temperature_K3,'gp', label="Temperature Sensor #3")
plt.plot(Voltage_V, line_fit3,'g-',label="")

plt.xlabel("Voltage [v]")
plt.ylabel("Temperature [K]")
plt.title('Temperature Calibration')
plt.legend()
plt.show()
```

### Just for fun, to see the effect of outliers on a best fit line, we will change Temperature_K3. In the graph above, it has an outlier of 307K at 16V, but if we move it to be more in line with the data, the best fit line would likely drop below the red line. Let's try it!

```python
Voltage_V = np.array([1.1,1.2,1.30,1.40,1.50,1.60,1.70,1.80,1.90])
Temperature_K3A = np.array([210, 190, 220, 235, 243, 237, 245, 261 , 278])

print('Voltage [V]: ', Voltage_V)
print('Temperature [K]: ', Temperature_K3A)
```

```python
# Generated linear fit
slope3, intercept3, r_value3, p_value3, std_err3 = stats.linregress(Voltage_V,Temperature_K3A)
line_fit3 = slope3 * Voltage_V + intercept3
```

```python
#randomly measured a voltage of 2.0352 V and figuring out what temperature it corresponds to

New_Voltage_V = 2.0352
New_Temp_K1 = New_Voltage_V * slope1
New_Temp_K2 = New_Voltage_V * slope2
New_Temp_K3A = New_Voltage_V * slope3

print('New Temperature Measurement 1 [K] :',New_Temp_K1)
print('New Temperature Measurement 2 [K] :',New_Temp_K2)
print('New Temperature Measurement 3 [K] :',New_Temp_K3A)
```

```python
plt.figure(figsize=(20,10))

plt.plot(Voltage_V,Temperature_K1,'b*', label="Temperature Sensor #1")
plt.plot(Voltage_V, line_fit1,'b--',label="")

plt.plot(Voltage_V,Temperature_K2,'rD', label="Temperature Sensor #2")
plt.plot(Voltage_V, line_fit2,'r--',label="")

plt.plot(Voltage_V,Temperature_K3A,'gp', label="Temperature Sensor #3")
plt.plot(Voltage_V, line_fit3,'g--',label="")

plt.xlabel("Voltage [v]")
plt.ylabel("Temperature [K]")
plt.title('Temperature Calibration')
plt.legend()
plt.show()
```

#### Just as predicted, the earlier outlier drastically changed the position of the best fit line!
